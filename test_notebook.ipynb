```python
import pandas as pd

def read_and_display_emails(filepath, num_emails=5):
    """
    Read the specified CSV file containing emails and display the full content of the first few emails.
    
    Parameters:
    - filepath: Path to the CSV file.
    - num_emails: Number of emails to display. Defaults to 5.
    """
    # Load the CSV file
    try:
        data = pd.read_csv(filepath)
        
        # Check if the 'message' column exists
        if 'message' in data.columns:
            # Display the specified number of emails
            for i in range(num_emails):
                print(f"Email {i+1} content:\n{data['message'].iloc[i]}\n")
                print("------------------------------------------------------------\n")
        else:
            print("The 'message' column does not exist in the provided CSV file.")
    except Exception as e:
        print(f"An error occurred while reading the file: {e}")

# Example usage
filepath = 'res_enron/emails.csv'
read_and_display_emails(filepath, num_emails=100)  # Adjust num_emails as needed
```
